```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Biotech_preprocessed_dataset_with_outliers.csv")
df['Time']=pd.to_datetime(df['Time'])
df
```

```python
start_date = pd.to_datetime('2024-03-01')
end_date = pd.to_datetime('2024-06-01')
df = df[(df['Time']>=start_date) & (df['Time']<end_date)]
df
```

```python
# generate a new dataframe containing columns for each hour
df_hours = pd.DataFrame(columns = [str(i) for i in range(24)])

# discard data after last date so that each column has equal number of rows
last_date = pd.to_datetime('2024-07-07 15:00:00')
df = df[df['Time']<last_date]

# copy the values of total power to corresponding hour columns of df_hours
for col in df_hours.columns:
    df_hours[col] = df[df['Hour']==int(col)]['Total (W)'].values

df_hours
```

```python
a = []
for col in df_hours.columns:
    # find mean and standard deviation
    mean = df_hours[col].mean()
    stdv = df_hours[col].std()
    size = 1000
    
     # generate 1000 normally distributed values 
    new_vals = np.random.normal(loc=mean, scale=stdv, size=1000)
    a.append(new_vals)

a = np.array(a)
a = a.T
a.shape 
```

```python
# generate a new dataframe of normally distributed values for each hour
hour_cols = [str(x) for x in range(24)]
df_hours_normal = pd.DataFrame(a, columns = hour_cols)
df_hours_normal
```

```python
df_hours_final = pd.concat([df_hours, df_hours_normal], ignore_index=True)
df_hours_final = df_hours_final.astype(int)
df_hours_final
```

```python
c = []
for i in range(1091):
    b = df_hours_final.iloc[i]
    b = b.tolist()
    c.extend(b)

df_final = pd.DataFrame({'Power':c})
df_final
```

```python
df_final.to_csv('normally_generated_seasoned_data.csv')
```
